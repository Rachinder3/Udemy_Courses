#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
### can use fillpoly function
#### instead of poly lines use fillpoly function
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_copy = img.copy()

vertices = [[225,700],[375,400],[600,700]]

pts = np.reshape(vertices, (-1,1,2))


cv2.fillPoly(img_copy,[pts], color=(0,0,255))
```

```python
plt.imshow(img_copy)
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python

### function

def draw_circle(event, x,y, flags, params):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x,y),radius = 100, color = (0,0,128),thickness=5)



img = cv2.imread(path)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) ## not needed as we are not using matplotlib and can stay in BGR


### linking

cv2.namedWindow(winname="dog")
cv2.setMouseCallback("dog",draw_circle)

### display image

while True:
    cv2.imshow("dog",img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
    
```
